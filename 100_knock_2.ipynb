```python
uriage["purchase_date"] = pd.to_datetime(uriage["purchase_date"])
uriage["purchase_month"] = uriage["purchase_date"].dt.strftime("%Y%m")
res = uriage.pivot_table(index="purchase_month",columns="item_name",aggfunc="size",fill_value=0)
res
```

```python
uriage["purchase_date"] = pd.to_datetime(uriage["purchase_date"])
uriage["purchase_month"] = uriage["purchase_date"].dt.strftime("%Y%m")
res2 = uriage.pivot_table(index="purchase_month",columns="item_name",values="item_price",aggfunc="sum",fill_value=0)
res2
```

```python
uriage["item_name"] = uriage["item_name"].str.upper()
uriage["item_name"] = uriage["item_name"].str.replace("　","")
uriage["item_name"] = uriage["item_name"].str.replace(" ","")
uriage.sort_values(by=["item_name"],ascending=True)
print(pd.unique(uriage["item_name"]))
```

```python
uriage.isnull().any(axis=0)
```

```python
flg_is_null = uriage["item_price"].isnull()
for trg in list(uriage.loc[flg_is_null,"item_name"].unique()):
    price = uriage.loc[(~flg_is_null) & (uriage["item_name"] == trg),"item_price"].max()
    uriage["item_price"].loc[(flg_is_null) & (uriage["item_name"] == trg)] = price
uriage.head()    
```

```python
uriage.isnull().any(axis=0)
```

```python
for trg in list(uriage["item_name"].sort_values().unique()):
    print(trg+"の最大額:"+str(uriage.loc[uriage["item_name"]==trg]["item_price"].max())+"の最小額:"+str(uriage.loc[uriage["item_name"]==trg]["item_price"].min(skipna=False)))
    
```
